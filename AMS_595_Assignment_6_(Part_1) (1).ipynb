## 2. Implementing Gradient Descent Algorithm (60 pts)

Given a function $f(x) = \sqrt{x^2+5}$, $x\in \mathbf{R}$.

Fact: This function is L-smooth.
0. Find the gradient of this function and the minimum value of this function analytically.

Use the last digit of your SB ID divided by 5 and then plus 1 as the initial guess (for example, if your ID ends with 9, your initial guess $x_0$ will be $9\div5 + 1= 2.8$).
3. Record the values of $x_k$ at the $k$-th iteration during GD and report $x_T$.

5. For each of the step sizes 5, 3, 1, and 0.5, perform gradient descent and record the values of $x_k$ in each step $k$. Plot $f(x_{k-1}) - f(x_{k})$ v.s. $k$ for each step size.
```python
# Plot the function f(x), np.linspace might be helpful.

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
func = np.sqrt(x**2 + 5)
plt.plot(x, func, label="f(x) = sqrt(x^2 + 5)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.scatter(0, np.sqrt(5), color='red', label='Minimum')
plt.legend()
plt.show()


```

```python
# Define the derivative of the function f(x), for gradient descent

def f(x):
  return np.sqrt(x**2 + 5)

def dfdx(x):
  return x / np.sqrt(x**2 + 5)


# write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.
def gradient_descent(x_0, alpha, num_iterations):
    x_k = [x_0]
    for i in range(num_iterations):
        x_next = x_k[-1] - alpha * dfdx(x_k[-1]) #takes last element of the list, subtracts step size times derivative
        x_k.append(x_next)
    return x_k

# set up/assign initial values
last_digit = 1
initial_guess = last_digit / 5 + 1
step_size = 1
T = 50

# Run gradient descent by calling the function
x_values = gradient_descent(initial_guess, step_size, T)

# report x_T by printing out the value
print("x_T:", x_values[-1])

# Plot the f(x_k) values vs. iteration number
f_values = [f(x) for x in x_values]
plt.figure(figsize=(12, 6))
plt.plot(range(T + 1), f_values, label="Step size α=1")
plt.xlabel("Iteration (k)")
plt.ylabel("f(xk)")
plt.title("f(xk) vs Iteration Number for α=1")
plt.legend()
plt.grid(True)
plt.show()


```

```python
# create a list containing step sizes
step_sizes = [5, 3, 1, 0.5]

# initialize a dictionary to store the results from different step sizes
results = []

# loop through the step sizes and perform GD and record results
for alpha in step_sizes:
    x_values = gradient_descent(initial_guess, alpha, T)
    f_differences = [f(x_values[k - 1]) - f(x_values[k]) for k in range(1, len(x_values))]  #find differences for each alpha
    results.append(f_differences)

# plot multiple curves in a single graph

plt.figure(figsize=(12, 6))
for alpha, x_values in enumerate(results):
    plt.plot(range(1, T + 1), results[alpha], label=f"Step size α={alpha}")


plt.xlabel("Iteration (k)")
plt.ylabel("f(xk-1) - f(xk)")
plt.title("f(xk-1) - f(xk) vs Iteration Number for Various Step Sizes")
plt.legend()
plt.grid(True)
plt.show()
```

**Report the parameter values and plot the generated data and the regression plane (in 1D, we have a line, but in 2D, we have a plane).** `numpy.linspace`, `numpy.meshgrid`, and [Surface Plot](https://matplotlib.org/stable/gallery/mplot3d/surface3d.html) methods might be helpful in plotting the plane.
```python
import numpy as np
import matplotlib.pyplot as plt

#generate synthetic data
np.random.seed(81)   #change the seed to be the last two digits of your SB ID.
X = 2 * np.random.rand(100, 2)  #generate 100 random 2D data points
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100)  #generate y values with noise

#add a column of ones to X for the bias terms (intercept) in the normal equation
X_ones = np.c_[np.ones((100, 1)), X]


#find theta using the normal equation
theta = np.linalg.inv(X_ones.T.dot(X_ones)).dot(X_ones.T).dot(y)


#print the intercept and coefficients of the regression plane
intercept = theta[0]
coefficients = theta[1:]
print("Intercept:", intercept)
print("Coefficients:", coefficients)

#plot the generated data and the regression plane

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], y, color='b', label='Data points') #plotting generated data

x0_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 10)
x1_vals = np.linspace(X[:, 1].min(), X[:, 1].max(), 10)
x0_mesh, x1_mesh = np.meshgrid(x0_vals, x1_vals) #create mesh grid for the surface plot

y_vals = intercept + coefficients[0] * x0_mesh + coefficients[1] * x1_mesh #calculate y predictions

ax.plot_surface(x0_mesh, x1_mesh, y_vals, color='r', alpha=0.5, label='Regression plane') #plot the linear regression plane

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
ax.set_title('Linear Regression Plane')

plt.show()

```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. You task is to
1. Tune the Learning Rate and Number of Epochs: Your first task is to fine-tune the values for the learning_rate and num_epochs parameters. Your goal is to identify suitable values that will enable us to converge to a set of parameter values $\theta$ closely approximating the optimal parameter values $\theta^*$.
2. Code Explanation: To enhance code comprehension, please augment the code with meaningful comments. These comments should elucidate the purpose and functionality of each code segment, making it easier for readers to understand the logistic regression implementation.

By accomplishing these tasks, we aim to achieve a better understanding of the logistic regression model's behavior and its parameter optimization process.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z): #the sigmoid function transforms the linear output of the model into a probability value (between 0 and 1)
    return 1 / (1 + np.exp(-z))


np.random.seed(595)
X = np.random.rand(2, 100) #generate 100 random samples for X
w_true = np.array([1.5, -2.5]) #define weights of the model
b_true = 1.0  #true bias
probabilities = sigmoid(np.dot(w_true.T, X) + b_true) #compute true probabilities
Y = (probabilities > 0.5).astype(int) #convert probabilities to binary (0 or 1)
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100) #add random noise to simulate real data

#split data into training and testing
X_train, X_test = X[:,:80], X[:,80:]
Y_train, Y_test = Y[:80], Y[80:]

#initialize weights and bias to 0
#learning rate 5.95 and num_epochs 595 gives accuracies .89 and .95
w = np.zeros(X.shape[0])
b = 0.0
learning_rate = 20 #after trial and error, 20 gives the highest accuracy
num_epochs = 2000 #after trial and error, 2000 gives the highest accuracy

for epoch in range(num_epochs):
    A_train = sigmoid(np.dot(w.T, X_train) + b) #calculate probabilities for the training set

    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train) #compute gradient of the loss function w.r.t. weights
    dJdb = np.mean(A_train - Y_train) #compute gradient of the loss function w.r.t. bias

    w -= learning_rate * dJdw
    b -= learning_rate * dJdb

A_train = sigmoid(np.dot(w.T, X_train) + b) #after training, compute probabilities for training set
predictions_train = (A_train > 0.5).astype(int) #convert to binary

A_test = sigmoid(np.dot(w.T, X_test) + b) #calculate probabilities for the test set
predictions_test = (A_test > 0.5).astype(int) #convert to binary

train_accuracy = np.mean(predictions_train == Y_train)

test_accuracy = np.mean(predictions_test == Y_test)

print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

```python
plt.scatter(X[0, :80], X[1,:80], c=Y[:80], cmap=plt.cm.Paired) #scatter plot for training data
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
ax = plt.gca()

xlim = ax.get_xlim() #retrieve x and y limits for the plot
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50)) #meshgrid allows us to evaluate the decision boundary over a continuous space
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b #compute the linear combination of features for each point
Z = sigmoid(Z) #get probabilities
Z = Z.reshape(xx.shape) #reshape into meshgrid
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--']) #plot the decision boundary

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```

```python
plt.scatter(X[0, 80:], X[1, 80:], c=Y[80:], cmap=plt.cm.Paired) #scatter plot for test data
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50)) #meshgrid allows us to evaluate the decision boundary over a continuous space
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b #compute the linear combination of features for each point
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--']) #plot the decision boundary

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```
